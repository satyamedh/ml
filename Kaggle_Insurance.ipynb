```python
import tensorflow as tf
import numpy as np
import random as rnd
import pandas as pd
import matplotlib.pyplot as plt

print(tf.test.gpu_device_name())
print(tf.__version__)
tf.random.set_seed(42)
```

```python
insurance = pd.read_csv("datasets/insurance.csv")
X_unprocessed = insurance.iloc[:,0:-1]
X_unprocessed
```

```python
y_unprocessed = insurance.iloc[:,-1:]
y_unprocessed
```

# Proccessing the labels(y)

```python
y_stage2 = y_unprocessed.values.tolist()
y_stage2
```

```python
y_stage3 = [i[0] for i in y_stage2]
y_stage3
```

```python
y_tobeused = y_stage3
```

# Proccessing the training data(X)

```python
def convert_smoking_or_no_to_one_or_zero(x):
    return (0 if x[0] == "no" else 1)
smoker_list = X_unprocessed.iloc[:,-2:-1].values.tolist()
for person in smoker_list:
    smoker_list[smoker_list.index(person)] = convert_smoking_or_no_to_one_or_zero(person)
    
smoker_list
```

```python
gender_column = X_unprocessed.iloc[:,-5:-4].values.tolist()
gender_column

def gender_function(x):
    return (0 if x[0] == "male" else 1)


gender = [gender_function(x) for x in gender_column]
gender
```

```python
region_column = X_unprocessed.iloc[:,-1:].values.tolist()
def region_function(x):
    return (0 if x[0] == "northeast" else (1 if x[0] == "northwest" else (2 if x[0] == "southeast" else (3 if x[0] == "southwest" else (None)))))
region = [region_function(x) for x in region_column]
region
```

```python
age = X_unprocessed.iloc[:,0:-5].values.tolist()
bmi = X_unprocessed.iloc[:,2:-3].values.tolist()
children = X_unprocessed.iloc[:,3:-2].values.tolist()
```

```python
X = []
y = []

X_test = []
y_test = []

for i in range(1002):
    X.append([age[i][0], gender[i], bmi[i][0], children[i][0], smoker_list[i], region[i]])
    y.append(y_tobeused[i])

for i in range(1003, 1338):
    X_test.append([age[i][0], gender[i], bmi[i][0], children[i][0], smoker_list[i], region[i]])
    y_test.append(y_tobeused[i])
```

# Creating the model!

```python
# Set random seed
tf.random.set_seed(42)


# CREATE MODEL!!!!!
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(7, activation="relu"))
model.add(tf.keras.layers.Dense(15, activation="relu"))
model.add(tf.keras.layers.Dense(25, activation="relu"))
model.add(tf.keras.layers.Dense(50, activation="relu"))
model.add(tf.keras.layers.Dense(75, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(75, activation="relu"))
model.add(tf.keras.layers.Dense(50, activation="relu"))
model.add(tf.keras.layers.Dense(25, activation="relu"))
model.add(tf.keras.layers.Dense(15, activation="relu"))
model.add(tf.keras.layers.Dense(7, activation="relu"))
model.add(tf.keras.layers.Dense(1))


# Compile model
model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.Adam(lr=0.0001), metrics=['mae', 'accuracy'])
```

# FIT THE MODEL!!!!!

```python
model.fit(X, y, epochs=10000)
```

# Check our model

```python
model.summary()
```

# Evaluate our model

```python
model.evaluate(X_test, y_test)
```
